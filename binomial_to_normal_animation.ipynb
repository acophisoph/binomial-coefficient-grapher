```
from matplotlib import pyplot as plt

# Python 3 code for Pascal's Triangle 
# A simple O(n^3) 
# program for 
# Pascal's Triangle 

# Function to print 
# first n lines of 
# Pascal's Triangle 
def printPascal(n) : 
	
    ### Create counter to make a list of x variables for each step
    count = []
    counter = 0
    # Iterate through every line 
    # and print entries in it
    for line in range(0, n) :

        ### Create list to append binmoial coefficients
        alllist = []
        # Every line has number of 
        # integers equal to line 
        # number 
        for i in range(0, line + 1) :
            '''
            print(binomialCoeff(line, i), 
                " ", end = "")
            '''
            alllist.append(binomialCoeff(line,i))
        ### Append list for numbers of coefficients at each step
        counter+=1
        count.append(counter)
        ### Create Binomial Distributiuon graphs (x-var, coefficients)
        fig = plt.figure()
        fig.set_figwidth(10+counter/10)
        plt.bar(count, alllist, counter/250)
        plt.show()

# See https://www.geeksforgeeks.org/space-and-time-efficient-binomial-coefficient/ 
# for details of this function 
def binomialCoeff(n, k) : 
    res = 1
    if (k > n - k) : 
        k = n - k 
    for i in range(0 , k) : 
        res = res * (n - i) 
        res = res // (i + 1) 
	
    return res 

# Driver program 
n = 250
printPascal(n) 


# This code is contributed by Nikita Tiwari. 
```
